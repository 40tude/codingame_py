## **1️⃣ Proportional (Kp) – Correction proportionnelle à l'erreur**
### **Effet :**  
- C'est le **terme principal** du PID.
- Il applique une correction proportionnelle à l'écart (`erreur = consigne - mesure`).
- Plus **Kp est élevé**, plus la réaction est **forte et rapide**.

### **Avantages :**
✅ Permet une réponse rapide.  
✅ Facile à régler.  

### **Inconvénients :**
❌ Trop grand → **Oscillations**, voire instabilité.  
❌ Trop petit → **Réaction lente**, atteint difficilement la cible.  

### **Dans ton cas** :
Avec **Kp = 4.0** seul, ton lander ajuste **directement la poussée** en fonction de l'erreur de vitesse.  
🚀 C'est efficace et simple car ton problème est principalement un contrôle de vitesse.


## **2️⃣ Integral (Ki) – Correction accumulée de l'erreur**
### **Effet :**
- Corrige une **erreur persistante** sur la durée.
- Accumule les erreurs passées (**intégrale de l'erreur**).
- Permet de **réduire l’erreur statique** qui persiste avec Kp seul.

### **Avantages :**
✅ Supprime **l'erreur résiduelle** (ex: si Kp seul ne suffit pas à atteindre -40 m/s).  
✅ Aide quand un système est soumis à une **force continue** (ex: gravité).  

### **Inconvénients :**
❌ Trop grand → **Effet mémoire**, instabilité, oscillations.  
❌ Trop petit → Pas d’effet notable.  

### **Pourquoi tu n’en as pas besoin ?**  
Ton problème est **instantané** (contrôle de vitesse), pas un problème avec une erreur persistante.  
📌 **Le vent, la gravité sont constants**, donc l'effet **intégral est inutile**.


## **3️⃣ Derivative (Kd) – Réduction des variations brusques**
### **Effet :**
- Réagit à la **vitesse de changement** de l'erreur.
- Prévient les **oscillations** en amortissant la correction.
- Plus Kd est grand, plus la réponse devient **fluide**.

### **Avantages :**
✅ **Anticipe** les erreurs, réduit les oscillations.  
✅ **Amortit** les surcorrections.  

### **Inconvénients :**
❌ Trop grand → Trop lent, empêche d’atteindre rapidement la cible.  
❌ Trop petit → Peu d'effet notable.  

### **Pourquoi tu n’en as pas besoin ?**  
Ton **réglage Kp = 4.0** fonctionne bien, sans oscillation visible.  
📌 **Kd est surtout utile pour des systèmes oscillants**, ce qui n’est pas ton cas ici.


## **Résumé de ton cas**
🎯 **Ton problème est un simple contrôle de vitesse**.  
🚀 **Kp seul suffit** pour converger rapidement à -40 m/s.  
📌 **Ki et Kd ne sont pas nécessaires** car il n'y a ni erreur résiduelle significative, ni oscillations.  


## **Quand utiliser chaque terme ?**
| **Type de problème**      | **Kp** | **Ki** | **Kd** |
|---------------------------|-------|-------|-------|
| Réponse rapide nécessaire | 🔥 Oui  | Non  | Parfois  |
| Correction d’une erreur persistante | Parfois | 🔥 Oui | Non |
| Éviter les oscillations | Parfois | Parfois | 🔥 Oui |
| Contrôle de vitesse simple | 🔥 Oui | Non | Non |
| Contrôle précis (ex: bras robotique) | Oui | Parfois | 🔥 Oui |

✅ Tu as bien fait de **supprimer Ki et Kd**.  
✅ **Kp = 4.0** est un bon compromis pour **réagir vite sans trop d’oscillations**.  
✅ **Si tu avais des oscillations**, ajouter **Kd** pourrait aider.
